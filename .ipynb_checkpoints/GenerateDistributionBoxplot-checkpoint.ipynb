# Code to generate the Boxplot with the distribution of all datasets

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

%matplotlib notebook
%matplotlib inline
```

```python
# Import all datasets
```

```python
df_S = pd.read_csv("data/Talus_Syene.csv", sep=';')
df_L = pd.read_csv("data/Talus_London_Soay.csv", sep=';')
df_K = pd.read_csv("data/Talus_Halle_Karakul.csv", sep=';')
df_M = pd.read_csv("data/Talus_Halle_Marsch.csv", sep=';')
df_C = pd.read_csv("data/Talus_Chicago.csv", sep=';')
```

```python
def clean_df(df, columns, target, name_male, name_female):
    X = df[columns].astype(float)
    X = pd.concat([X, df[target]], axis=1)
    X = X.dropna(axis=0).reset_index(drop=True)
    X = X.replace(to_replace=name_male, value='1').replace(to_replace=name_female, value='0').replace(to_replace='c', value='1')

    label = X[target]

    X = X.drop(target, axis=1)
    
    return X, label
```

```python
col_talus = ['GLl', 'GLm', 'Dl', 'Bd']

X_S, y_S = clean_df(df_S, col_talus, ['Sex'], 'm', 'f')
X_L, y_L = clean_df(df_L, col_talus, ['Sex'], 'm', 'f')
X_K, y_K = clean_df(df_K, col_talus, ['Sex'], 'm', 'f')
X_M, y_M = clean_df(df_M, col_talus, ['Sex'], 'm', 'f')
X_C, y_C = clean_df(df_C, col_talus, ['Sex'], 'm', 'f')
```

```python
X_S = X_S.assign(Dataset='Syene')
X_L = X_L.assign(Dataset='London')
X_K = X_K.assign(Dataset='Karakul')
X_M = X_M.assign(Dataset='Marsch')
X_C = X_C.assign(Dataset='Chicago')
```

```python
plt.rcParams['figure.figsize'] = [7, 6]

df = pd.concat([X_S, X_L, X_K, X_M, X_C], axis=0)
#print(df.head())
mdf = pd.melt(df, id_vars=['Dataset'], var_name=['Measurement'])
#print(mdf.head())
fig1, ax1 = plt.subplots()
ax = sns.boxplot(x="Measurement", y="value", hue="Dataset", data=mdf, palette="deep", fill=False, gap=.1)
ax.yaxis.grid(True)
ax.xaxis.grid(True)

fig1.savefig('Boxplot-all.pdf', bbox_inches='tight')
```
