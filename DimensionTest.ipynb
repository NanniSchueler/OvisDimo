# Tests on which Feature has the most impact

```python
import numpy as np
from itertools import combinations
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
```

```python
import pandas as pd

file_path = 'data/Talus_Syene.csv'  
data = pd.read_csv(file_path, sep=';') 

print(data.head())
```

```python
data['RA'] = data['GLl'] / data['GLm']
data['RA'].replace([np.inf, -np.inf], np.nan, inplace=True)

features = ['Bd', 'Dl', 'GLm', 'GLl', 'RA']
label = 'Sex'

feature_combinations = []
for r in range(1, 6):
    feature_combinations.extend(combinations(features, r))

results = []

for combo in feature_combinations:
    subset = data[list(combo) + [label]].dropna()
    X = subset[list(combo)]
    y = subset[label]

    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    svm = SVC(kernel='linear')
    
    accuracy = cross_val_score(svm, X_scaled, y, cv=5, scoring='accuracy').mean()
    results.append({'Features': combo, 'Accuracy': accuracy})

results_Syene = pd.DataFrame(results)

#results_Syene.to_csv('ratioSyeneResult.csv', index=False)
```

```python
import pandas as pd

file_path = 'data/Talus_London_Soay_noC.csv'  
data1 = pd.read_csv(file_path, sep=';') 

print(data1.head())
```

```python
data1['RA'] = data1['GLl'] / data1['GLm']
data1['RA'].replace([np.inf, -np.inf], np.nan, inplace=True)

features = ['Bd', 'Dl', 'GLm', 'GLl', 'RA']
label = 'Sex'

feature_combinations = []
for r in range(1, 6):
    feature_combinations.extend(combinations(features, r))

results = []

for combo in feature_combinations:
    subset = data1[list(combo) + [label]].dropna()
    X = subset[list(combo)]
    y = subset[label]

    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    svm = SVC(kernel='linear')
    
    accuracy = cross_val_score(svm, X_scaled, y, cv=5, scoring='accuracy').mean()
    results.append({'Features': combo, 'Accuracy': accuracy})

results_Soay = pd.DataFrame(results)
#results_Soay.to_csv('ratio_LondonResult.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_data(data, title):
    data.sort_values(by='Accuracy', ascending=False, inplace=True)

    #data['Features'] = data['Features'].str.replace(',', '\n', regex=False)
    data['Features'] = data['Features'].apply(lambda x: '\n'.join(x))

    plt.figure(figsize=(12, 8))  
    plt.plot(data['Features'], data['Accuracy'], marker='o')
    plt.title(title)
    plt.xlabel('Features')
    plt.ylabel('Accuracy')
    plt.xticks(rotation=0)  
    plt.tight_layout() 
    plt.show()

plot_data(results_Soay, 'Dimension Test London Results')
#plt.savefig('LondonResults.pdf', bbox_inches='tight')
```

```python
plot_data(results_Syene, 'Dimension Test Syene Results')
#plt.savefig('SyeneResults.pdf', bbox_inches='tight')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_combined_data(data1, data2, title):
    data1.sort_values(by='Accuracy', ascending=False, inplace=True)
    #data1['Features'] = data1['Features'].apply(lambda x: '\n'.join(x))

    #data2['Features'] = data2['Features'].apply(lambda x: '\n'.join(x))
    
    mapping = dict(zip(data2['Features'], data2['Accuracy']))
    data2_mapped = data1['Features'].map(mapping)

    plt.figure(figsize=(12, 3.5))
    plt.plot(data1['Features'], data1['Accuracy'], marker='o', label='SVM Results Syene ')
    plt.plot(data1['Features'], data2_mapped, marker='X', label='SVM Results Soay')
    plt.title(title)
    plt.xlabel('Features')
    plt.ylabel('Accuracy')
    plt.xticks(rotation=0)
    plt.legend(loc='lower left') 
    plt.tight_layout()
    plt.savefig('CombinedResults.pdf', bbox_inches='tight')
    plt.show()
    
plot_combined_data(results_Syene, results_Soay, '')
```
