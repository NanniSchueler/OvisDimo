```python
def norm_df(df, columns):
    df = df.dropna(axis=0).reset_index(drop=True)
    
    min_params = df[columns].min()
    max_params = df[columns].max()

    df[columns] = (df[columns] - min_params) / (max_params- min_params)
    return df
```

```python
needs = ['GLl', 'GLm', 'Dl', 'Bd', 'Sex']
columns = ['GLl', 'GLm', 'Dl', 'Bd']

df_norm_syene = norm_df(df_syene[needs], columns)

df_norm_soay = norm_df(df_soay[needs], columns)
df_norm_karakul = norm_df(df_karakul[needs], columns)
df_norm_marsch = norm_df(df_marsch[needs], columns)

df_norm_kara = norm_df(df_kara_clean[columns], columns)

print(f"Length of the original dataframe: {len(df_kara_clean)} - new length: {len(df_norm_kara)}")
```

```python
def clean_df(df, columns, target, name_male, name_female):
    X = df[columns].astype(float)
    X = pd.concat([X, df[target]], axis=1)
    X = X.dropna(axis=0).reset_index(drop=True)
    X = X.replace(to_replace=name_male, value='1').replace(to_replace=name_female, value='0').replace(to_replace='c', value='1')

    label = X[target]

    X = X.drop(target, axis=1)
    
    return X, label
```

```python
col_talus = ['GLl', 'GLm', 'Dl', 'Bd']

df_rezent = pd.concat([df_norm_soay, df_norm_karakul, df_norm_marsch], axis=0).reset_index(drop=True)
#df_alt = df_norm_syene

X_rez, y_rez = clean_df(df_rezent, col_talus, ['Sex'], 'm', 'f')
X_alt, y_alt = clean_df(df_norm_syene, col_talus, ['Sex'], 'm', 'f')

df_rezent.head()
```
